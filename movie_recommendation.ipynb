```python
#https://www.kaggle.com/tmdb/tmdb-movie-metadata
credits = pd.read_csv("tmdb_5000_credits.csv")
movies_df = pd.read_csv("tmdb_5000_movies.csv")
```

```python
credits_column_renamed = credits.rename(index=str,columns={"movie_id": "id"})
movies_df_merge =movies_df.merge(credits_column_renamed, on='id')
movies_df_merge.head()
```

```python
movies_cleaned_df =movies_df_merge.drop(columns=['homepage', 'title_x','title_y', 'status','production_countries'])
movies_cleaned_df.head()
movies_cleaned_df.info()
```

```python
# Calculate all the components based on the above formula
v=movies_cleaned_df['vote_count']
R=movies_cleaned_df['vote_average']
C=movies_cleaned_df['vote_average'].mean()
m=movies_cleaned_df['vote_count'].quantile(0.70)
movies_cleaned_df['weighted_average']=((R*v)+(C*m))/(v+m)
movies_cleaned_df.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
movies = pd.read_csv('tmdb_5000_movies.csv')

# Calculate average and threshold
C = movies['vote_average'].mean()
m = movies['vote_count'].quantile(0.70)

# Filter and calculate weighted average
qualified = movies[movies['vote_count'] >= m].copy()

def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v / (v + m)) * R + (m / (m + v)) * C

qualified['weighted_average'] = qualified.apply(weighted_rating, axis=1)

# Sort by weighted average
movie_sorted_ranking = qualified.sort_values('weighted_average', ascending=False)

# Now sort by popularity
popularity = movie_sorted_ranking.sort_values('popularity', ascending=False)

# Show top 5
print(popularity.head())

# Optional: Plot top 10 by popularity
plt.figure(figsize=(12,6))
sns.barplot(
    x=popularity['popularity'].head(10),
    y=popularity['original_title'].head(10),
    data=popularity
)
plt.title("Top 10 Movies by Popularity")
plt.xlabel("Popularity Score")
plt.ylabel("Movie Title")
plt.show()
```
