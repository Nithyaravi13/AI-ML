```python
import pandas as pd
import numpy as np
```

```python
s1=pd.Series(np.random.randint(1,high=5,size=100, dtype='I'))
s2=pd.Series(np.random.randint(1,high=4,size=100, dtype='I'))
s3=pd.Series(np.random.randint(10000,high=30001,size=100, dtype='I'))
print(s1,s2,s3)
```

```python
housemkt = pd.concat([s1,s2,s3], axis=1)
housemkt.head()
```

```python
housemkt.rename(columns={0: 'bedrs', 2: 'price_sqr_meter'}, inplace=True)
housemkt.head()
```

```python
#join concat the values
bigcolumn = pd.concat([s1,s2,s3], axis=0)

# it is still a series , so we need to transform it to a dataframe
bigcolumn = bigcolumn.to_frame()
print(type(bigcolumn))

bigcolumn
```
