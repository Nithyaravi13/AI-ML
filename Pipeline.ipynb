```python
from sklearn.pipeline import Pipeline
##feature Scaling
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
steps=[("standard_scaler",StandardScaler()),
       ("classifier",LogisticRegression())]
```

```python
steps
```

```python
pipe=Pipeline(steps)
```

```python
from sklearn.impute import SimpleImputer
```

```python
## numerical processing pipeline
import numpy as np
numeric_processsor = Pipeline(
    steps=[("imputation_mean",SimpleImputer(missing_values=np.nan, strategy="mean")),
           ("scaler", StandardScaler())]


)
```

```python
numeric_processsor
```

```python
##categorical processing pipeline
```

```python
from sklearn.preprocessing import OneHotEncoder
categorical_processsor=Pipeline(
    steps=[("imputation_constant",SimpleImputer(fill_value="missing",strategy="constant")),
           ("onehot",OneHotEncoder(handle_unknown="ignore"))]
)
```

```python
categorical_processsor
```

```python
##combine processing techniques
from sklearn.compose import ColumnTransformer
```

```python
preprocessor=ColumnTransformer(
    [("categorical",categorical_processsor,["gender","City"]),
     ("numerical",numeric_processsor,["age","height"])]

)
```

```python
preprocessor
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe=make_pipeline(preprocessor,LogisticRegression())
```
